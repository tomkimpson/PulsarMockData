# Compare frequency data.

This notebook compares the data produced for different input parameters 


```python
file_1 = '../output_data/canonical/frequency_timeseries.npz'
file_2 = '../output_data/canonical_low_noise/frequency_timeseries.npz'
file_2 = '../output_data/canonical_super_low_noise/frequency_timeseries.npz'
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def process_data_file(path,num_to_plot):

    data = np.load(path)
    y = data['f_tim_file'] 

    t = y[:,-1]
    tplot = t / (365*24*3600)
    plt.figsize=(12,12)
    for i in range(num_to_plot):
        plt.plot(tplot,y[:,i])

    plt.show()
    return y[:,:num_to_plot]
```

```python
num_to_plot = 3
y1 = process_data_file(file_1,num_to_plot)
y2 = process_data_file(file_2,num_to_plot)
```

These look very similar since they have the same seed. But the GW amplitude is different so they should not be identical! 

Lets check:

```python
plt.plot(y1-y2)
```
